# Variable Descriptions

Below is a list of variables commonly found in wage or labor market panel data sets, along with brief explanations of each.

---

## exp
Likely “Experience.” Often measured as “potential experience,” for example:  

\[
\text{exp} = \text{age} - \text{education} - 6
\]  
(assuming age at starting school is 6), or as self-reported years of work experience.


## wks
“Weeks worked” in the year, i.e., how many weeks the individual was employed during the survey year.

## occ
“Occupation” category, often coded numerically to represent different occupations (e.g., Managerial, Professional, Clerical, etc.).

## ind
“Industry” category, coded numerically to represent the industry sector (e.g., Manufacturing, Construction, Services, etc.).

## south
A dummy variable (0 or 1) indicating whether the individual resides in the Southern region of the United States.  
- 1 = South  
- 0 = Non-South

## smsa
A dummy variable indicating residence in a Standard Metropolitan Statistical Area (SMSA).  
- 1 = In SMSA  
- 0 = Not in SMSA

## ms
“Marital status.” Could be a dummy variable (e.g., 1 = married, 0 = otherwise) or a categorical code for multiple statuses. Refer to specific documentation for details.

## fem
A dummy variable for “female.”  
- 1 = Female  
- 0 = Male

## union
A dummy variable indicating union membership.  
- 1 = Union member  
- 0 = Non-union

## ed
“Education,” typically measured in years of schooling. For example:  
- 12 = High school graduate  
- 16 = College graduate  
- etc.

## blk
A dummy variable for race, typically indicating “Black” or “African American.”  
- 1 = Black  
- 0 = Otherwise

## lwage
“Log of wage.” The natural logarithm of the individual’s wage (hourly, weekly, or annual, depending on the dataset).

## id
A unique individual identifier. Each person in the panel has a distinct `id`.

## t
“Time” or “Year” variable. In a panel dataset, this identifies the wave or year of observation (e.g., 1, 2, 3, or actual years like 1976, 1977, etc.).

## tdum1, tdum2, ..., tdum7
“Time dummy” variables for different time periods.  
For example:  
- `tdum1 = 1 if t=1, 0 otherwise`  
- `tdum2 = 1 if t=2, 0 otherwise`  
- and so on.

These dummies capture year-specific effects in the panel data regression.

## exp2
“Experience squared,” i.e. \(\text{exp}^2\). This is often included in wage equations to allow for a non-linear (usually diminishing) return to experience.

```python
import pandas as pd
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
```

```python
# Activate the automatic pandas to R data frame conversion
pandas2ri.activate()

ro.r('install.packages("plm", repos="https://cloud.r-project.org")')

# Make sure these R packages are installed: install.packages("plm")
ro.r('library(plm)')
```

```python


# 1) Suppose df_py is your pandas DataFrame with columns:
#    id, t, lwage, wks, union, exp, exp2, south, smsa, fem, blk, ed
# (adjust to your real data)

# df_py = ... # your real data
# For demonstration, let's assume df_py is already loaded.

# 2) Convert Python DataFrame to an R data.frame
r_df_name = "df_r"  # Name in R environment
ro.globalenv[r_df_name] = pandas2ri.py2rpy(df_py)

# 3) Convert to panel data structure in R
#    index=c("id","t") tells plm which columns are individual/time
ro.r(f"{r_df_name} <- plm.data({r_df_name}, index=c('id','t'))")

# 4) Construct the R formula for Hausman–Taylor
#    NOTE: Adjust variable names to exactly match your columns in df_py
r_formula = """
lwage ~ wks + union + exp + exp2 + south + smsa + fem + blk + ed
    | wks + union + exp + exp2 + fem + blk
"""

# 5) Fit the Hausman–Taylor model
#    random.method="ht" and model="random" can also be used,
#    but "model='ht'" in recent plm versions typically suffices.
#    Alternatively: model="random", random.method="ht"
r_command = f"""
ht_model <- plm(
    formula = {r_formula},
    data = {r_df_name},
    model = 'ht'
)
"""

ro.r(r_command)

# 6) Retrieve the summary in R
summary_cmd = "ht_summary <- summary(ht_model)"
ro.r(summary_cmd)

# 7) Pull the summary object back to Python
ht_summary = ro.globalenv['ht_summary']

```

```python
# You can display the summary in text form:
print(ht_summary)

# Or extract coefficients from the summary:
# The summary object typically has a 'coefficients' matrix in R:
coefficients_matrix = ht_summary.rx2('coefficients')
print(coefficients_matrix)

# If you want them as a Python DataFrame:
df_coefs = pd.DataFrame(coefficients_matrix)
print(df_coefs)
```
